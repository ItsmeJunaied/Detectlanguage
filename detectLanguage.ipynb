```python
import os
import csv
from google.cloud import translate_v2 as translate

# Set the environment variable for Google Cloud service account key
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/content/ultra-depot-381621-ecd1a99c28c1.json'

# Instantiates a client
translate_client = translate.Client()

# Define a function to detect the language of a sentence
def detect_language(text):
    # Detect the language
    detection = translate_client.detect_language(text)

    # Return the detected language
    return detection['language']

# Read the input CSV file
with open('finalBIGGESTDATA3 - Sheet1.csv', mode='r') as input_file:
    reader = csv.reader(input_file)

    # Create a new CSV file for the output
    with open('outputetaiFinal.csv', mode='w', newline='') as output_file:
        fieldnames = ['Sentence', 'Language']
        writer = csv.DictWriter(output_file, fieldnames=fieldnames)
        writer.writeheader()

        # Loop through each row in the input file
        for row in reader:
            sentence = row[0]
            language = detect_language(sentence)
            writer.writerow({'Sentence': sentence, 'Language': language})
            print(f'Sentence: {sentence}, Language: {language}')
```
